```python
events = pd.read_csv('../data/events.csv')
# events.event.value_counts()
events = events[['visitorid','itemid','event']][:10000]
print("Pre Shape",events.shape)
events['rating'] = events['event'].apply(lambda x: 1 if x=='view' else 2 if x=='addtocart' else 3 if x=='transaction' else null)
events = events.sort_values('rating').drop_duplicates(['visitorid','itemid'], keep='last')[['visitorid','itemid','rating']]
print("Post Shape",events.shape)
```

```python
items1 = pd.read_csv('../data/item_properties_part1.csv')
items2 = pd.read_csv('../data/item_properties_part2.csv')
items = pd.concat([items1, items2])
items = items[
    (items.itemid.isin(events.itemid))
    & (items.property == 'categoryid')
]
items_lkp = (items[['itemid','value']]
             .drop_duplicates()
             .groupby(['itemid'], as_index=False)
             .agg(','.join)
            )
items_lkp[items_lkp['value'].str.contains(',')==True].head(10)
```

```python
interactions = create_interaction_matrix(df = events,
                                         user_col = 'visitorid',
                                         item_col = 'itemid',
                                         rating_col = 'rating')
interactions.head()
```

```python
# Create User Dict
user_dict = create_user_dict(interactions=interactions)
# Create Item dict
items_dict = create_item_dict(df = items_lkp,
                               id_col = 'itemid',
                               name_col = 'itemid')
```

```python
mf_model = runMF(interactions = interactions,
                 n_components = 30,
                 loss = 'warp',
                 epoch = 30,
                 n_jobs = 4)
```
